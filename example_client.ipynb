```python
# assumes that docker container is running
import requests
import json

local_guess_endpoint = "http://localhost:8080/api/guesses"

# Example request data (adjust based on your API requirements)
data = {
    "team_id": "team123",
    "guesses": ["halts", "afoul"]
}

# true answers: afoul, after

try:
    # Make POST request
    response = requests.post(local_guess_endpoint, json=data)

    # Print response details
    print(f"Status Code: {response.status_code}")
    print(f"Response Headers: {response.headers}")
    print(f"Response Body:")
    print(json.dumps(response.json(), indent=2))

except requests.exceptions.ConnectionError:
    print("Error: Could not connect to localhost:8080")
    print("Make sure the server is running")
except requests.exceptions.JSONDecodeError:
    print(f"Status Code: {response.status_code}")
    print(f"Response Text: {response.text}")
except Exception as e:
    print(f"Error: {e}")
```
